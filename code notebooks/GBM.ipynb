```python
df = pd.read_csv('snote.csv')
```

```python

# Extract features (all columns except the target variable)
X = df.drop(columns=['Label'])

# Extract target variable (label)
y = df['Label']
```

```python
# Binarize the target variable for class 0 vs. all others
y_binary = (y == 0).astype(int)
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training (70%) and test (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Further split the training set into training (70%) and validation (30%) sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=42)

# Now have to X_train, X_val, X_test for features and y_train, y_val, y_test for labels
```

```python
#2 Gradient boosting machine algorithm

xgb_classifier = xgb.XGBClassifier(random_state=42)


xgb_classifier.fit(X_train, y_train)

# y_pred = xgb_classifier.predict(X_test)
```

```python
# Make predictions 
y_pred_gbm = xgb_classifier.predict(X_test)
```

```python
# Compute confusion matrix for each classifier

from sklearn.metrics import confusion_matrix

cm_gbm = confusion_matrix(y_test, y_pred_gbm)

print(cm_gbm)

```

```python
# Compute TPR, TNR, FPR, FNR for each classifier using confusion matrix
def compute_metrics(cm):
    TP = cm[1, 1]
    TN = cm[0, 0]
    FP = cm[0, 1]
    FN = cm[1, 0]

    tpr = TP / (TP + FN)
    tnr = TN / (TN + FP)
    fpr = FP / (FP + TN)
    fnr = FN / (FN + TP)

    return tpr, tnr, fpr, fnr

# Compute metrics for each classifier
tpr_gbm, tnr_gbm, fpr_gbm, fnr_gbm = compute_metrics(cm_gbm)

print(compute_metrics(cm_gbm))
```

```python
from sklearn.metrics import accuracy_score

# Compute accuracy for each classifier

accuracy_gbm = accuracy_score(y_test, y_pred_gbm)

```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision_gbm = precision_score(y_test, y_pred_gbm, average='weighted')
recall_gbm = recall_score(y_test, y_pred_gbm, average='weighted')
f1_gbm = f1_score(y_test, y_pred_gbm, average='weighted')

```

```python
# Print computed metrics for each classifier


print("Gradient Boosting Machines Metrics:")
print("Precision:", precision_gbm)
print("Recall:", recall_gbm)
print("F1 Score:", f1_gbm)
print()
```
